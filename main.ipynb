```python
# import training data and store it in a dataframe using pandas

import pandas as pd

WIFI_training_data = pd.read_csv('trainingData.csv')

# head() is used to see first five rows of the dataframe
WIFI_training_data.head()
```

```python
# print shape of training dataframe

print(WIFI_training_data.shape)
```

```python
# import validation data and store it in a dataframe using pandas

WIFI_validation_data = pd.read_csv('validationData.csv')
# print shape of validation dataframe
print(WIFI_validation_data.shape)
```

```python
# drop columns which dont attribute to the final prediction 

WIFI_training_data_dropCol = WIFI_training_data.drop(columns = ['LONGITUDE', 'LATITUDE', 'SPACEID', 'USERID', 'PHONEID', 'TIMESTAMP', 'RELATIVEPOSITION'] )

WIFI_training_data_dropCol.head()

# we can also select longitude and latitude as the target variables but we will go with buildingID and floor 
```

```python
# drop columns from validation dataset as well

WIFI_validation_data_dropCol = WIFI_validation_data.drop(columns = ['LONGITUDE', 'LATITUDE', 'SPACEID', 'USERID', 'PHONEID', 'TIMESTAMP', 'RELATIVEPOSITION'] )
WIFI_validation_data_dropCol.head()
```

```python
# drop null values if any

WIFI_training_data_dropCol.dropna(axis=0)
print(WIFI_training_data_dropCol.shape)
WIFI_validation_data_dropCol.dropna(axis=0)
print(WIFI_validation_data_dropCol.shape)
```

```python
#printing X_train which contains the entire training data

X_train = WIFI_training_data_dropCol.iloc[:,0:520]
X_train
```

```python
# printing y_train with two target variables: floor and buildingID

y_train = WIFI_training_data_dropCol.iloc[:,520:]
y_train
```

```python
#printing X_validation which contains the entire training data

X_validation = WIFI_validation_data_dropCol.iloc[:,0:520]
X_validation
```

```python
# printing y_validation with two target variables: floor and buildingID

y_validation = WIFI_validation_data_dropCol.iloc[:,520:]
y_validation
```

```python
# this scaling function converts all 100s to -510 and then adds all values with 105. 
# After this min-max scaling is applied column wise so that all values range between 0 and 1

def min_max_scaling(df):
    # copy the dataframe
    df_norm = df.copy()
    # apply min-max scaling
    
    for column in df_norm.columns:
        df_norm[column]=df_norm[column].replace(100, -510)
        df_norm[column] = df_norm[column] + 105
#         100 -> -510 and -510 + 105 -> -405
#         0 to -104 -> 105 to 1
        df_norm[column] = (df_norm[column] - df_norm[column].min()) / (df_norm[column].max() - df_norm[column].min() + 1)
    return df_norm

# apply normalization to X_train

X_train_normalized = min_max_scaling(X_train)
X_train_normalized
```

```python
# apply normalization to X_validation

X_validation_normalized = min_max_scaling(X_validation)
X_validation_normalized
```

```python
# created a function for calling random forrest 

def RFC(X, y):
    rfc = RandomForestClassifier(n_estimators = 600, max_depth = None).fit(X, y)
    y_pred = rfc.predict(X_validation_normalized)
    return accuracy_score(y_validation_new, y_pred)
```

```python
# import train_test_split

from sklearn.model_selection import train_test_split

#divide data  into 80% and 20% and also set the random_state to 0, 1, 2 to get random samples everytime

X_80_0, X_20_0, y_80_0, y_20_0 = train_test_split(X_train_normalized, y_train_new, test_size = 0.2, random_state = 0) 
X_80_1, X_20_1, y_80_1, y_20_1 = train_test_split(X_train_normalized, y_train_new, test_size = 0.2, random_state = 1) 
X_80_2, X_20_2, y_80_2, y_20_2 = train_test_split(X_train_normalized, y_train_new, test_size = 0.2, random_state = 2) 

# calling RFC function defined above to check accuracy score for random forrest

accuracy_score_20_0 = RFC(X_20_0, y_20_0)
accuracy_score_20_1 = RFC(X_20_1, y_20_1)
accuracy_score_20_2 = RFC(X_20_2, y_20_2)

# calculating average for 20%

accuracy_score_20_avg = (accuracy_score_20_0 + accuracy_score_20_1 + accuracy_score_20_2) / 3

# calling RFC function defined above to check accuracy score for random forrest

accuracy_score_80_0 = RFC(X_80_0, y_80_0)
accuracy_score_80_1 = RFC(X_80_1, y_80_1)
accuracy_score_80_2 = RFC(X_80_2, y_80_2)

# calculating average for 80%

accuracy_score_80_avg = (accuracy_score_80_0 + accuracy_score_80_1 + accuracy_score_80_2) / 3

#divide data  into 60% and 40% and also set the random_state to 0, 1, 2 to get random samples everytime

X_60_0, X_40_0, y_60_0, y_40_0 = train_test_split(X_train_normalized, y_train_new, test_size = 0.4, random_state = 0) 
X_60_1, X_40_1, y_60_1, y_40_1 = train_test_split(X_train_normalized, y_train_new, test_size = 0.4, random_state = 1) 
X_60_2, X_40_2, y_60_2, y_40_2 = train_test_split(X_train_normalized, y_train_new, test_size = 0.4, random_state = 2) 

# calling RFC function defined above to check accuracy score for random forrest

accuracy_score_40_0 = RFC(X_40_0, y_40_0)
accuracy_score_40_1 = RFC(X_40_1, y_40_1)
accuracy_score_40_2 = RFC(X_40_2, y_40_2)

# calculating average for 40%

accuracy_score_40_avg = (accuracy_score_40_0 + accuracy_score_40_1 + accuracy_score_40_2) / 3

# calling RFC function defined above to check accuracy score for random forrest

accuracy_score_60_0 = RFC(X_60_0, y_60_0)
accuracy_score_60_1 = RFC(X_60_1, y_60_1)
accuracy_score_60_2 = RFC(X_60_2, y_60_2)

# calculating average for 60%

accuracy_score_60_avg = (accuracy_score_60_0 + accuracy_score_60_1 + accuracy_score_60_2) / 3

# randomising data for 100% training data
X_100_0 = X_train_normalized.sample(frac=1, random_state=0).reset_index(drop=True)
y_100_0 = y_train_new.sample(frac=1, random_state=0).reset_index(drop=True)


X_100_1 = X_train_normalized.sample(frac=1, random_state=1).reset_index(drop=True)
y_100_1 = y_train_new.sample(frac=1, random_state=1).reset_index(drop=True)


X_100_2 = X_train_normalized.sample(frac=1, random_state=2).reset_index(drop=True)
y_100_2 = y_train_new.sample(frac=1, random_state=2).reset_index(drop=True)

# calling RFC function defined above to check accuracy score for random forrest

accuracy_score_100_0 = RFC(X_100_0, y_100_0)
accuracy_score_100_1 = RFC(X_100_1, y_100_1)
accuracy_score_100_2 = RFC(X_100_2, y_100_2)

# calculating average for 100%

accuracy_score_100_avg = (accuracy_score_100_0 + accuracy_score_100_1 + accuracy_score_100_2) / 3
```

```python
# plotting accuracy score for 20, 40, 60, 80 and 100% training data

import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")

training_data_percent = [20, 40, 60, 80, 100]
accuracy_score_list = [accuracy_score_20_avg, accuracy_score_40_avg, accuracy_score_60_avg, accuracy_score_80_avg, accuracy_score_100_avg]

sns.lineplot(training_data_percent, accuracy_score_list)
plt.title('Training data percent Vs Accuracy Score')
plt.xlabel('Training data')
plt.ylabel('Accuracy Score')
plt.show()
```
